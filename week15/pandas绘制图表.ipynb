```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', **{'family':'Microsoft Yahei, SimHei'}) # 加载
```

```python
data = pd.read_table("./data/cities.txt", sep=' ')
```

```python
data.head()
```

```python
data['平均工资'].plot()
plt.show()
```

```python
data = data.set_index('城市')
data.head()
```

```python
data['平均工资'][:10].plot(kind = 'bar') # 画柱状图
plt.show()
```

```python
data[['平均工资']].boxplot() # 画箱线图
plt.show()
```

```python
data[['平均工资']].hist() # 画箱线图
plt.show()
```

```python
import numpy as np

data = pd.DataFrame(np.random.rand(10,4), columns=list('ABCD'))
data
```

```python
data.plot(kind='bar') # 柱状图
plt.show()
```

```python
data.plot(kind='barh') # 横向柱状图
plt.show()
```
