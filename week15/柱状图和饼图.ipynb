```python
import matplotlib.pyplot as plt
```

```python
x = ['Q1','Q2','Q3','Q4']
y = [10,30,20,60]

rect = plt.bar(x,y,color='g', width=0.3) # width是调整宽度

for ind, item in enumerate(rect):
    _x = item.get_x() + 0.1
    _y = item.get_height() + 1
    plt.text(_x,_y,y[ind]) # 给每个轴写上相应的数字

# plt.text(0.2, 50, 'test') # 在0.2和50之间写test字
plt.ylim(0,70) # y轴的高度值

plt.grid(True) # 画网格
plt.show()
```
