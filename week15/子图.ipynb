```python
import matplotlib.pyplot as plt
```

# 子图

```python
plt.subplot(2,1,1)
# 传入的参数，第一个第二个为2行1列，第三个为第1个图形
plt.subplot(2,1,2)
plt.show()
```

```python
plt.subplot(3,1,1)
# 3行1列
plt.subplot(3,1,2)
plt.subplot(3,1,3)
plt.show()
```

```python
plt.subplot(1,2,1)
# 1行2列
plt.subplot(1,2,2)
plt.show()
```

```python
plt.subplot(2,2,1)
# 2行2列
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,1,2)
plt.show()
```

```python
import numpy as np

plt.subplot(2,2,1)

x = np.random.random(10)
y = np.random.random(10)
plt.scatter(x,y)
plt.title('test') # 要给每个图表写标题，在相应位置写上即可

plt.subplot(2,2,2)

x = np.arange(7)
y = x ** 2
plt.plot(x,y)
plt.title('test2')

plt.subplot(2,1,2)

plt.show()
```
