# 异常值处理

```python
df.loc[6] = {'name':'test','age':999,'score':'C'}
```

```python
df
```

```python
q_upper = df['age'].quantile(0.75)
q_upper
```

```python
q_lower = df['age'].quantile(0.25)
q_lower
```

```python
val = q_upper - q_lower
val
```

```python
k = 1.5
```

```python
df[df['age'] > q_upper + k * val]
```

```python
df2 = df.drop(5)
```

```python
df2
```

```python
df[(df['age'] < q_upper + k * val) & (df['age'] > q_upper - k * val)]
```
