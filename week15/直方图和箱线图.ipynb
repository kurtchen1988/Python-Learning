```python
import matplotlib.pyplot as plt
import numpy as np
```

# 直方图

```python
np.random.seed(100)
data = np.random.normal(9000,3000, size=300)
data = data[data>=4000]
data
```

```python
len(data)
```

```python
plt.hist(data)
plt.show()
```

```python
plt.hist(data, color='g', rwidth=0.8, alpha = 0.6)
# alpha是两条之间的距离，默认是1就是没有空格
# rwidth是每条的宽度
plt.show()
```

```python
plt.hist(data, bins = 20)
# bins是数量的意思
plt.show()
```

```python
len(data[data<4900])
```

```python
len(data[(data>7000) & (data<9500)])
```

# 箱线图

```python
data = np.arange(1,10)
data
```

```python
plt.boxplot(data)
# 上下界线是箱图的平均值上下的1.5倍
plt.show()
```

```python
np.random.seed(100)
data = np.random.normal(size=100)
data
```

```python
plt.hist(data)
plt.show()
```

```python
data = np.concatenate([data, [4,7,8,9,-4]]) # 这里的数据超出了非常多范围，对于数据分析是比较大的干扰。
#数据分析的时候应该剔除这些数据。
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data, whis=3)
plt.show()
```

```python
plt.boxplot(data, showmeans = True, meanline=True)
# showmeans是显示中位值，meanline是显示中位值线
plt.show()
```

```python
plt.boxplot(data, showbox=False)
# showbox是控制显示框
plt.show()
```
