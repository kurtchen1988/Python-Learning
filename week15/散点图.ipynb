```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
x
```

```python
np.random.seed(1)
y = np.random.random(size=10)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='r', linestyle='--', marker='o')
# linestyle是线的样式，--为虚横线，:为虚点，-.是虚线加点
# marker是将点画出来
plt.show()
```

```python
plt.plot(x,y,'b--*')
# 可以将所有样式写在一起
plt.show()
```

```python
x = np.linspace(0,10,100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y) # 得到平滑曲线的方法，让两个数之间得到更多的点即可
plt.show()
```

```python
y2 = np.cos(x)
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

```python
plt.plot(x,y,label='sinx(x)') # 给图案打上标签
plt.plot(x,y2,label='con(x)')
plt.legend()

plt.xlabel("x axis")
plt.ylabel('y value') # 给图案的x，y轴打上标签

plt.title('My Demo') # 写出图案的标题

plt.show()
```

```python
plt.plot(x,y,label='sinx(x)') # 给图案打上标签
plt.plot(x,y2,label='con(x)')
plt.legend()

plt.axis([5,10,0,1.2]) # 缩小取值范围

plt.show()
```
